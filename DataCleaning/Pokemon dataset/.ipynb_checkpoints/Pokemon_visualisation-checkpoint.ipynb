```python
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd 
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('pokemon.csv',encoding='utf-8',index_col=0)
```

```python
df.head()
```

```python
sns.lmplot(x="Attack",y="Defense",data=df,fit_reg=False,hue="Generation")
plt.xlim(0,None)
plt.ylim(0,None)
plt.show()
```

```python
sns.boxplot(data=df)
plt.show()
```

```python
df_stats = df.drop(['Legendary','Generation'],axis=1)
sns.boxplot(data=df)
plt.show()
```

```python
sns.set_style('whitegrid')
sns.violinplot(x='Type 1',y='Attack',data=df)
plt.show()
```

```python
sns.swarmplot(x='Type 1',y='Attack',data=df)
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.violinplot(x='Type 1',y='Attack',data=df,inner=None)
sns.swarmplot(x='Type 1',y='Attack',data=df,color='k',alpha=0.7)
plt.title('Attack by type')
plt.show()
```

```python
df_stats.head()
```

```python
df_melted=pd.melt(df_stats,id_vars=["Name","Type 1","Type 2"],var_name="Stat")
df_melted.head()
```

```python
plt.figure(figsize=(15,10))
sns.swarmplot(x="Stat",y="value",data=df_melted,hue="Type 1")
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.swarmplot(x="Stat",y="value",data=df_melted,hue="Type 1",dodge=True)
plt.legend(bbox_to_anchor=(1, 1), loc=2)
plt.show()
```

```python
#using Heatmaps 
corr = df_stats.corr()
sns.heatmap(corr)
plt.show()
# print(corr)
```

```python
sns.distplot(df.Attack)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.countplot(x="Type 1",data=df)
plt.show()
```

```python

g=sns.factorplot(x='Type 1',y="Attack",data=df,hue="Generation",col="Generation",kind="swarm")
g.set_xticklabels(rotation=-45)
plt.show()
```

```python
sns.kdeplot(df.Attack,df.Defense)
plt.show()
```

```python
sns.jointplot(x="Attack",y="Defense",data=df)
plt.show()
```
